```python
import numpy as np
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

data = np.loadtxt('E_Field.txt', skiprows=3)


# Extraindo as colunas
frequencia = data[:, 0]  # Frequncia em PHz
campo_eletrico = data[:, 1]  # Magnitude do campo eltrico

# Converter frequncia (PHz) para comprimento de onda (nm)
# 1 PHz = 10^15 Hz, c = 299792458 m/s -> c / (f * 10^15)
comprimento_onda = 299792458 / (frequencia * 1e15) * 1e9  # Resultado em nm

# Normalizar o campo eltrico
campo_eletrico_normalizado = campo_eletrico / np.max(campo_eletrico)

# Plotar o grfico
plt.figure(figsize=(10, 6))
plt.plot(comprimento_onda, campo_eletrico_normalizado, label="Campo Eltrico Normalizado", color="blue")
plt.title("Grfico do Campo Eltrico (Normalizado) em Funo do Comprimento de Onda")
plt.xlabel("Comprimento de Onda (nm)")
plt.ylabel("Campo Eltrico (Normalizado)")
plt.grid(True)
plt.legend()
plt.show()
```
